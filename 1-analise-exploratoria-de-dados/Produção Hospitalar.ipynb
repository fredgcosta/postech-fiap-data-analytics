```python
import pandas as pd #1ª célula
from matplotlib import ticker
from matplotlib import pyplot as plt
```

```python
dados = pd.read_csv('../dados/sih_cnv_qiuf165241189_61_97_210.csv', encoding='ISO-8859-1', skiprows=3, sep=';', skipfooter=12, thousands='.', decimal=',', engine='python') #2ª célula
```

```python
dados["2008/Ago"]
```

```python
dados.plot(x="Unidade da Federação", y="2008/Ago")
```

```python
axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")
plt.show()
```

```python
axis = dados.plot(x="Unidade da Federação", y="2008/Ago", kind="bar", figsize=(9,6))
axis.yaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.2f}"))
plt.title("Valor por unidade da federação")
plt.xticks(rotation=75, ha='right')
plt.show()
```
